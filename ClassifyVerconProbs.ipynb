```python
# Clean/Normalize Arabic Text
def clean_str(text):
    search = ["أ","إ","آ","ة","_","-","/",".","،"," و "," يا ",'"',"ـ","'","ى","\\",'\n', '\t','&quot;','?','؟','!']
    replace = ["ا","ا","ا","ه"," "," ","","",""," و"," يا","","","","ي","",' ', ' ',' ',' ? ',' ؟ ',' ! ']
    
    #remove tashkeel
    p_tashkeel = re.compile(r'[\u0617-\u061A\u064B-\u0652]')
    text = re.sub(p_tashkeel,"", text)
    
    #remove longation
    p_longation = re.compile(r'(.)\1+')
    subst = r"\1\1"
    text = re.sub(p_longation, subst, text)
    
    text = text.replace('وو', 'و')
    text = text.replace('يي', 'ي')
    text = text.replace('اا', 'ا')
    
    for i in range(0, len(search)):
        text = text.replace(search[i], replace[i])
    
    #trim    
    text = text.strip()

    return text
```

```python
from sklearn.cross_validation import train_test_split
train,_test,train_labels,test_labels = train_test_split(dat["P_DESCRIPTION"],                   
                                                 dat['PROBLEM_CAT_ID'], test_size=0.25,
                                                 random_state=1)

print ('        len: ' , len(train))
print (train[:5])
print (train_labels[:5])
print(len(train))
print(len(_test))
```

```python
from sklearn import naive_bayes
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn import metrics


# In[267]:


#It is important to note that the CountVectorizer performs lowercasing by default. 
vectorizer = CountVectorizer(lowercase=False).fit(train)
print (len(vectorizer.get_feature_names()))
print (vectorizer.get_feature_names()[-300:])
```

```python

train_vectorized = vectorizer.transform(train)


# In[269]:


test_vectorized = vectorizer.transform(_test)


# In[271]:


#create a naive bayes classifier 
clfr = naive_bayes.MultinomialNB()
clfr.fit(train_vectorized,train_labels)
predicted = clfr.predict(test_vectorized)
acc = metrics.accuracy_score(test_labels,predicted)
print ('accuracy = '+str(acc*100)+'%')
print (metrics.classification_report(test_labels,predicted))
```
